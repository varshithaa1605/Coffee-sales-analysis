```python
import matplotlib.pyplot as plt
```

```python
xy.columns
```

```python
a=xy['Total_Bill']
plt.plot(a)
plt.show()
```

```python
category=xy['product_category'].value_counts()
b=category.values
c=category.index
clr=['skyblue','lavender','lightpink','lightgreen']
plt.figure(figsize=(10,15),dpi=100)
plt.title("Catergory")
plt.pie(b,labels=c,autopct='%1.1f%%',colors=clr)
plt.legend(c,loc='upper right')
plt.show()
```

```python
name=xy['product_type'].value_counts()
d=name.values
e=name.index
plt.figure(figsize=(25,15),dpi=100)
plt.title("NAMES")
plt.pie(d,labels=e,autopct='%1.1f%%')
plt.show()
```

```python
f=xy[xy['unit_price']>3]['Size'].value_counts()
g=f.index
h=f.values
plt.title("Orders on Coffee size whose price is geater than ")
plt.xlabel("Size")
plt.ylabel("Counts")
plt.bar(g,h,color='skyblue')
plt.show()
```

```python
popular=xy.groupby('product_detail')['transaction_qty'].sum().head(10).sort_values()
popular.plot(kind='bar')
plt.show()
```

```python
xy.columns
```

```python
revenue=xy.groupby('product_category')['Total_Bill'].sum().sort_values()
revenue.plot(kind='barh',color='lavender')
plt.show()
```

```python
i=xy['unit_price'].value_counts().sort_index()
plt.title("Distribution of unit prize")
plt.xlabel("Unit price")
plt.ylabel("Counts")
plt.scatter(i.index,i.values)
plt.grid()
plt.show()
```

```python
j = xy[xy['Size'] == 'Large']
k = j['product_category'].value_counts()
l = k.values
m = k.index
plt.title("Coffee Category Distribution for Large Size")
plt.pie(l, labels=m, autopct='%1.1f%%', colors=['gray', 'lavender', 'black'])
plt.legend(m, loc='upper right')
plt.show()
```
